### Q1: What is the maximum and minimum mpg for Ford and Chevy cars with more than 150 HP?

```python
# One Options
car_makes = list(set(list(df.Auto)))
#car_makes = df.Auto.unique() # Alternatively

ford = [x for x in car_makes if 'Ford' in x]
chevy = [x for x in car_makes if 'Chev' in x]

ford_chevy_cars = ford + chevy

ford_chevy_cars
```

```python
ford_chevy_df = df[df['Auto'].isin(ford_chevy_cars)]
ford_chevy_df
```

```python
# Another Option
ford_chevy_df = df[(df.Auto.str.contains('Ford') | df.Auto.str.contains('Chev'))]
ford_chevy_df
```

```python
ford_chevy_df['HP'].max()
```

```python
ford_chevy_df.HP.max()
```

There are **no** records with an HP > 150 since the **max** in our dataframe is equal to 130.0

What about records with > 120 HP?

```python
meet_hp_reqs = ford_chevy_df[ford_chevy_df.HP > 120] # 3 left
meet_hp_reqs
```

```python
min_ = meet_hp_reqs['MPG'].min() # 17
max_ = meet_hp_reqs['MPG'].max() # 19.2

print('The min = {} and the max = {}'.format(min_, max_))
```

```python
# and Another Option

min_ = df[(df.Auto.str.contains('Ford') | df.Auto.str.contains('Chev')) & (df.HP > 120)]['MPG'].min()
max_ = df[(df.Auto.str.contains('Ford') | df.Auto.str.contains('Chev')) & (df.HP > 120)]['MPG'].max()

print('The min = {} and the max = {}'.format(min_, max_))
```

```python
# and yet Another Option

min_ = df[(df.Auto.str.contains('(Ford|Chev)')) & (df.HP > 120)]['MPG'].min()
max_ = df[(df.Auto.str.contains('(Ford|Chev)')) & (df.HP > 120)]['MPG'].max()

print('The min = {} and the max = {}'.format(min_, max_))
```

### Q4: What is the average MPG for each year for 6 cylinder cars? Do not include 1974. Or Fords.

```python
# No yearly data; calculate in aggregate
new_df = copy.deepcopy(df)
new_df['Ford_Filter'] = new_df['Auto'].apply(lambda x: 1 if 'Ford' in x else 0)
new_df
```

```python
mean_w_ford = new_df['MPG'].mean() # 24.76052631578947
mean_wo_ford = new_df[new_df['Ford_Filter'] == 0]['MPG'].mean() # 25.27647058823529

print('The average MPG including Ford is {}'.format(mean_w_ford))
print('The average MPG excluding Ford is {}'.format(mean_wo_ford))
```
